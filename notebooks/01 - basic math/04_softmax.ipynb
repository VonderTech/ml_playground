## Softmax
- The natural exponent, e, is approximately equal to 2.71828. It's a mathematical constant that is the base of the natural logarithm. It's called "natural" because it comes up naturally in many areas of mathematics.

```python
# f(x) = e^x

import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return np.exp(x)


x = np.linspace(-3, 2, 100)
y = f(x)
plt.plot(x, y)
plt.show()
```

## What does softmax do?
- The softmax function is used in machine learning, specifically in the field of neural networks. It's a type of squashing function, which is used to normalize the output of a network to a probability distribution over predicted output classes.
- It's often used in the final layer of a neural network model, where we're interested in outputting a probability distribution over classes. 

```python
# example of calculating softmax manually

import numpy as np
import matplotlib.pyplot as plt


numbers = np.random.randint(-5, 5, size=20)
# print(numbers)
# calculate the exponential of each number
exp_values = np.exp(numbers)
# calcualte the sum of all exponential values
sum_exp_values = np.sum(exp_values)
# calculate the probabilities
probabilities = exp_values / sum_exp_values
# print(probabilities)

plt.title("$\sum\sigma$ = %g" % np.sum(probabilities))
plt.plot(numbers, probabilities, "o")
plt.show()
```

Using PyTorch

```python
import torch
import torch.nn as nn

softfun = nn.Softmax(dim=0)
numbers = np.random.randint(-5, 5, size=20)

sigmaT = softfun(torch.tensor(numbers, dtype=torch.float32))
plt.plot(numbers, sigmaT, "o")
```
